Ques-3 Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

- The Above Five function are the main fuction to make a data in readable format and they make data easy to use.

- The Trim function is use to remove unwanted spaces between words in a specific cell.

 - The Left Function is used to pickup or to get number of words from the left side of a specific Name or from something else.

 - The Right Function is used to pickup or to get number of words from the Right side of a specific Name or from something else.

 - The Mid Function is used to pickup or to get number of words from a Name or something but it enables from which word we need to pickup or how much of words we need to extract from that Name or from something else.

 - The Fuction Concat is use to join or to add two or more name , sentances or other things so that data can be clear concise and make it in readable format.
